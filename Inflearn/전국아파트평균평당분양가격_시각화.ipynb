## `.str.replace("a", "b")` 데이터 내부 빈칸제거

* 문자열에서 "a" -> "b" 로 변경

* 연도와 월 분리하기

## `df[ ].apply(function)` 으로 데이터값 처리하고 가져오기

````python
date = "2013년12월"
# 연 반환 함수
def parse_year(date):
    year = date.split("년")[0]
    year = int(year)
    return year

# 월 반환 함수
def parse_month(date):
    month = date.split("년")[-1].replace("월","")
    month = int(month)
    return month

parse_year(date), parse_month(date)
````

````python
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt.head(1)
````
