```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams['xtick.labelsize'] = 10
import 
```

```python
train_df = pd.read_csv("data/train_df.csv")
train_df.tail(3)
```

```python
# class 종류와 counts
train_df['class'].value_counts()
```

```python
classlists = train_df['class'].unique()
```

```python
# label 종류 개수와 counts
# length = 88
train_df['label'].value_counts()
```

```python
for classlist in classlists:
    temp_df = train_df[train_df["class"] == classlist][["class","label"]].groupby("label").count().rename(columns={"class":"count"}).sort_values(by='count', ascending=False)
    fig, axs = plt.subplots(figsize=(15, 5))
    axs.bar(temp_df.index, temp_df.iloc[:, 0])
    axs.set_title(classlist, fontsize=20)
    for j, value in enumerate(temp_df.iloc[:, 0]):
        axs.text(j, 20, temp_df.iloc[:, 0][j], ha='center', fontsize=20)
    plt.show()
```
